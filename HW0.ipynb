```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Make the column playerID in the csv file as the index column and the first row as the header. Also, skip the second row when reading the file.
import pandas as pd
df_salaries_data = pd.read_csv('Salaries.csv', index_col = "playerID", header = 0, skiprows = [1])
print(type(df_salaries_data))
print(df_salaries_data)
```

```python
#Select the id of the players who are registered in ATL and HOU and whose salary is higher than one million.
selected_players_index = df_salaries_data[(df_salaries_data["teamID"].isin(["ATL","HOU"])) & (df_salaries_data["salary"] > 100000)].index
print(selected_players_index)
```

```python
# Use the describe() method to calculate the standard deviation, first quartile, median, third quartile, mean, maximum, and minimum of the salary in team ATL.
salary_ATL_team = df_salaries_data[df_salaries_data["teamID"] == "ATL"]["salary"].describe()
salary_ATL_team
```

```python
## convert to dic
## iterrows() method\
dic_salaries_data = {'playerID':[],'yearID':[], 'teamID':[], 'lgID':[], 'salary':[]} 
for index, row in df_salaries_data.iterrows():
    dic_salaries_data['playerID'].append(index)  
    dic_salaries_data['yearID'].append(row['yearID'])
    dic_salaries_data['teamID'].append(row['teamID'])
    dic_salaries_data['lgID'].append(row['lgID'])
    dic_salaries_data['salary'].append(row['salary'])

print(dic_salaries_data)
```

```python
## convert to dic
## to_dict method\
# my_dict = df_salaries_data.to_dict()
my_dict_df = pd.DataFrame(df_salaries_data.to_dict())
my_dict_df
```

```python
# Create a 2-dimensional Python list object, then convert it to a Numpy array object.
list1 = [[1,3,5],[2,4,6]]
arr = np.array(list1)
arr
```

```python
# Examine the ndim, shape, size, dtype, itemsize, and data attributes of the numpy array object.
print("nidm:",arr.ndim) #no. of axes (dimension) of the array
print("shape:",arr.shape) #(n,m): n=row, m=column
print("size:",arr.size) #total no. of elements of the array
print("dtype:",arr.dtype) #type of data in array
print("itemsize:",arr.itemsize) #the size in bytes of each element of the array np.float64 np.int32 np.float32
print("data attributes:",arr.data) #the buffer containing the actual elements of the array
```

```python
# Learn the dimension concept of an ndarray object by using reshape() and flatten() methods.
arr2= arr.reshape(3,2)
arr3 = arr.flatten()
print("reshape:",arr2)
print("flatten:",arr3)
```

```python
# slice practice
arr4=[[1,2,3],[4,5,6],[7,8,9]]
arr4=np.array(arr4)
print("arr4:\n", arr4)
print("row 2:\n", arr4[2])
print("row 1 & 2:\n", arr4[:-1])
print("column 2:\n", arr4[:,1])
print("column 1 & 3:\n", arr4[:,::2])
print("n in intersection of row 2 and column 2,3", arr4[1,1:])
```

```python
# argmin(), argmax(), min(), max(), mean(), sum(), std(), dot(), square(), sqrt(), abs(). exp(), sign(), and mod() methods.
print("argmin:",arr4.argmin()) # 
print("argmax:",arr4.argmax()) # 
print("min:",arr4.min())
print("max:",arr4.max())
print("mean:",arr4.mean())
print("sum:",arr4.sum())
print("std:",arr4.std())
arr5 = np.array([[-10,0,12],[13,-14,15],[16,17,-18]])
print("dot:",np.dot(arr4, arr5)) # 
print("square:",np.square(arr4)) # 
print("sqrt:",np.sqrt(arr4)) # 
print("abs:",np.abs(arr5)) # 
print("exp:",np.abs(arr4)) # e
print("sign:",np.sign(arr5)) #  0 1 0 -1 0 0
print("mod:",np.mod(arr5, 2)) #

```

```python
# Examine the arange(), ones(), zeros(), eye(), linspace(), and concatenate() methods.
arr6 = np.arange(1, 15, 3)
print("arange:", arr6)
arr7 = np.ones((3,4)) # 1 
print("ones:", arr7)
arr8 = np.zeros((2,3)) # 0 
print("zero:", arr8)
arr9 = np.eye(3,4,k=1)  #  1 0k 0 
print("eye:", arr9)
arr10 = np.linspace(0,18,4) # 
print("linspace:", arr10)
v_concat = np.concatenate([arr4, arr5], axis=0)
print("v_concat:", v_concat) # 
h_concat = np.concatenate([arr4, arr5], axis=1)
print("h_concat:", h_concat) # 
```

```python

x = np.linspace(-5, 5, num=200)
y = np.array([j ** 2 for j in x])

plt.plot(x,y)
plt.title("Parabola")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis label")
plt.grid()
plt.show()
```

```python
x = np.linspace(-5, 5, num=20)
y1 = np.array([j for j in x])
y2 = np.array([j ** 2 for j in x])
y3 = np.array([j ** 3 for j in x])

plt.plot(x, y1, label='linear')
plt.legend(loc="best")
plt.text(-0.5, 3, '$y=x$',color='red')
plt.show()

plt.plot(x, y2, label='quadratic')
plt.legend(loc="best")
plt.text(-5.5, 3, '$y=x^2$', fontsize=20)
plt.show()
        
plt.plot(x, y3, label='cubic')
plt.legend(loc="best")
plt.text(-0.5, 3, 'y=x^3', fontsize=20)
plt.show()
```
